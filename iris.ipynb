```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import adaline
from matplotlib.colors import ListedColormap
```

```python
s = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

df = pd.read_csv(s, header=None, encoding='utf-8')
```

```python
# Get the first 100 rows and four coluns
y = df.iloc[0:100, 4].values
# Filter
y = np.where(y=='Iris-setosa', 0 , 1)
```

```python
X = df.iloc[0:100, [0,2]].values
```

```python
# PLot values
plt.scatter(X[:50, 0], X[:50, 1], color='red', marker='o', label='Setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color='blue', marker='s', label='Veriscolor')
plt.xlabel('Sepal length [cm]')
plt.ylabel('Petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```

```python
# Applying standardization feature scaling. 
mean_x1 = np.mean(X[:, 0])
mean_x2 = np.mean(X[:, 1])
std_x1 = np.std(X[:, 0])
std_x2 = np.std(X[:, 1])

X[:,0] = X[:,0] - mean_x1 / std_x1
X[:,1] = X[:,1] - mean_x2 / std_x2
```

```python
ada = adaline.Adaline(X, y)

ada.fit(300)
```

```python
plt.plot(ada.losses)
print(ada.losses)
```

```python
ada.test(X,y)
```
